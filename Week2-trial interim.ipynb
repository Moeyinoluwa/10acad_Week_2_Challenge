```python
from pandas import DataFrame
import matplotlib.pyplot as plt
Data = {'Variables': ['Minimum Value','Maximum Value','Average Value','1st Quartile','2nd Quartile','3rd Quartile'],
        'Values': [12.000000 , 18.000000 , 14.700000 , 12.250000 , 15.000000 , 16.750000]}
df = DataFrame(Data,columns=['Variables','Values'])
df.plot(x ='Variables', y='Values', kind = 'bar')
plt.xticks(rotation=30, horizontalalignment="center")
plt.title("Barplot showing the Minimum value, the Maximum value, the Average, the 1st, 2nd & 3rd Quartile for the number of xDR sessions")
plt.xlabel("Variables")
plt.ylabel("Corresponding Values")
plt.plot()
```
