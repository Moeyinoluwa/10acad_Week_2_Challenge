## Sessions frequency

```python
sesscount= sess[['Bearer Id']].count().sort_values(by= 'Bearer Id', ascending= False) #Session bearer id count for the top 10 id
sesscount.head(10)
```

### Session duration

```python
sess_dur= sess[['Dur. (ms)']].sum()
sess_dur.head(10)
```

```python
#Data description after grouping
to_analyze= sess.agg({
    'Dur. (ms)': 'sum',
    'Bearer Id': 'count',
    'Total DL (Bytes)': 'sum',
    'Total UL (Bytes)': 'sum'
})
to_analyze.describe().astype('int64')
```

## Loading some sklearn standard libraries

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
scale_model= StandardScaler()
```

```python
model_sc= scale_model.fit_transform(data[['Total DL (Bytes)','Total UL (Bytes)']])
```

```python
model= KMeans(n_clusters= 5)
```

```python
train_x= model.fit_predict(model_sc)
```

```python
data['Group data']= train_x
```

```python
data.head()
```

```python
data['Group data'].unique()
```

```python
to_analyze= data.pivot_table(index= 'Group data', values= ['Dur. (ms)','Bearer Id'])
```

```python
to_analyze
```

## Minimum, Maximum, Average & Total non- normalized metrics for each cluster

```python
model_describe= data.groupby('Group data')[['Dur. (ms)','Bearer Id']].describe().T
model_describe
```

## Visualization for the descriptions of each cluster

```python
plt.figure(figsize= (14, 9))
sns.heatmap(model_describe.xs('Dur. (ms)'), annot= True, fmt= '.2f')
plt.yticks(fontsize= 14)
plt.title('Dur. (ms) description');
```

```python
plt.figure(figsize= (14, 9))
sns.heatmap(model_describe.xs('Bearer Id').loc[['count', 'min', 'max'], :], annot= True, fmt= '.0f')
plt.yticks(fontsize= 14)
plt.title('Bearer ID Description');
```
